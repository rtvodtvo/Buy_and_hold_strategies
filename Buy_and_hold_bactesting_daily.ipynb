```python
dfs = [dca_result, seasonal_dca_result, above_ema_dca_result, under_ema_dca_result, dynamic_ema_dca_result, volatility_adjusted_dca_result]
df_names = ['dca_result', 'seasonal_dca_result', 'above_ema_dca_result', 'under_ema_dca_result', 'dynamic_ema_dca_result', 'volatility_adjusted_dca_result']
```

```python
df_best = dca_result.loc[['Total Return','Max Strategy Drawdown [%]','Max Strategy Drawdown Duration(Days)']]
df_best= df_best.drop(columns=['average'])
df_best.loc['Total Return'] = -20000
df_best.loc['Max Strategy Drawdown [%]'] = 20000
df_best.loc['Max Strategy Drawdown Duration(Days)'] = 200000
```

```python
#best values
for ticker in tickers:
    for df in dfs:
        if df[ticker]['Total Return'] > df_best[ticker]['Total Return']:
            df_best[ticker]['Total Return'] = df[ticker]['Total Return'] 
        if df[ticker]['Max Strategy Drawdown [%]'] <df_best[ticker]['Max Strategy Drawdown [%]']:
            df_best[ticker]['Max Strategy Drawdown [%]'] =  df[ticker]['Max Strategy Drawdown [%]']
        if df[ticker]['Max Strategy Drawdown Duration(Days)']<df_best[ticker]['Max Strategy Drawdown Duration(Days)']: 
            df_best[ticker]['Max Strategy Drawdown Duration(Days)'] = df[ticker]['Max Strategy Drawdown Duration(Days)']
        
```

```python
df_best
```

```python
#names of the strategies with best values
df_strategys = df_best
for ticker in tickers:
    for i in range(len(dfs)):
        if dfs[i][ticker]['Total Return'] ==  df_strategys[ticker]['Total Return']:
            df_strategys[ticker]['Total Return'] = df_names[i]
        if dfs[i][ticker]['Max Strategy Drawdown [%]'] == df_strategys[ticker]['Max Strategy Drawdown [%]']:
            df_strategys[ticker]['Max Strategy Drawdown [%]'] = df_names[i]
        if dfs[i][ticker]['Max Strategy Drawdown Duration(Days)'] == df_strategys[ticker]['Max Strategy Drawdown Duration(Days)']: 
            df_strategys[ticker]['Max Strategy Drawdown Duration(Days)']= df_names[i]
```

```python
df_strategys
```

```python
merged_df = df_strategys
for i in range(3):
    for j in range(9):
        merged_df.iloc[i,j] = df_strategys.iloc[i,j] + ' : '+ df_best.iloc[i,j].astype(str)
merged_df
```
